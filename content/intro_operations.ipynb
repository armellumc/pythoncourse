# Week 1: Workshop - Operations on variables

Adding variables to python scripts is a lot of fun, but the purpose of adding variables is to do something with their values. For instance, to add two numbers, to change some text, to read a list of records from a dataset, to export an excel sheet, to make plots etc. 

To do this we need to use those variables in more complex instructions. We go step by step and we start with operators:


| Operator  | Syntax | Example  | How to use in Python      |
|-----------|--------|----------|---------------------------|
| Add       | +      | a + b    | total = a + b            |
| Subtract  | -      | a - b    | total = a - b            |
| Divide    | /      | a / b    | total = a / b            |

ℹ️ **Note**: total is the name of a variable, that means you can choose another name and it will still work. For instance, my_result = a + b.

```python
a = 10
b = 15

print(a + b)
print(a - b)
print(a * b)      
print(a / b)
```
