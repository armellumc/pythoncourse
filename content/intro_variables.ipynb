### How do we know if a variable is numeric or text?

We use the type() function.

For each variable, print their type in python using type(), don’t forget to use print so Python will show the result of type():
Like this print(type(a))
Check: Do you get the following output in your terminal?
-	int – for print(type(a))

Then you’re good! If not, check your syntax in python. Are using print(type(VARIABLE)) for instance.
The type of a variable matters when you start using operators or compare variables, some combinations between numbers and text are not possible, or lead to some surprising results in Python. 
One reason is logical, what would for instance “hello” + 5 mean? It seems impossible to add a number (5) to a text value “hello” and find a result. 10 + 5 works. Those are two integers and Python knows that the result will be 15.
Another reason is that operators can be redefined in Python. This is very obvious with text values. While you are used to see operations like 10 + 5 and so on, you might not have seen something like “hello” + “ how are you doing?”. It also works, why? Because Python will just add the second sentence to the first one, making one sentence out of it. The value will be hello how are you doing?


```python
my_text = "hello"
my_other_text = " how are you doing?"
print(my_text + my_other_text)
```
